```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import numpy as np
import pandas as pd
import torch
import requests
from bs4 import BeautifulSoup
import re
```

```python
tokenizor = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
def sentiment_score(review):
    tokens = tokenizor.encode(review, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
sample  = con_data["tweet"]
```

```python
con_data.shape
```

```python
results=[]
for i in range(252):
    a = sentiment_score(sample[i])
    results.append(a)
```
